```python
import cv2

def count_frames(vid_path):
    capture = cv2.VideoCapture(vid_path)

    if not capture.isOpened():
        print("Error with opening file")
        return 0

    fps = capture.get(cv2.CAP_PROP_FPS)

    frame_count = 0
    while True:
        ret, _ = capture.read()
        if not ret:
            break
        frame_count += 1
    total_frames = frame_count

    capture.release()

    print(f"FPS: {fps}, Total frames: {total_frames}")

    return total_frames

count_frames('/kaggle/input/casme2/CASME2_compressed-20250812T164409Z-1-001/CASME2_compressed/sub02/EP01_11f.avi')
```
